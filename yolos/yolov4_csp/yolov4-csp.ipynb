# image、label

```python
import json

# bbox : x, y, w, h
with open('./data/annotations/instances_val2017.json', 'r') as f:
    d = json.load(f)
labels = d['annotations']
```

```python
from tqdm import tqdm
import os
import cv2

images_path = os.listdir('./data/val2017/')
for label in tqdm(labels):
    image_id = str(label['image_id'])
    
    p = '0' * (12 - len(image_id)) + image_id
    label_p = p + '.txt'
    img_p = os.path.join('./data/val2017/', p + '.jpg')
    h, w, _ = cv2.imread(img_p).shape
    
    category_id = str(label['category_id'])
    
    i1, i2, i3, i4 = label['bbox'][0], label['bbox'][1], label['bbox'][2], label['bbox'][3]
    
#     i1, i2, i3, i4 = label['bbox'][0] / w, label['bbox'][1] / h, label['bbox'][2] / w, label['bbox'][3] / h
#     i1, i2 = i1 - i3 / 2, i2 - i4 / 2
#     i3, i4 = i1 + i3, i2 + i4
    bbox = category_id + ' ' + str(i1) + ' ' + str(i2) + ' ' + str(i3) + ' ' + str(i4) + '\n'

    imp = os.path.join('./data/val2017/', label_p)
    with open(imp, 'a') as f:
        f.write(bbox)
```
